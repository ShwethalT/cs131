```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
import scipy.stats as stats

#opening the file and creating dataframe
with open("winequality-white.csv", "r") as f:
  whitedf = pd.read_csv("winequality-white.csv", sep = ";")

whitedf
```

# Question 1.

```python
whitedf.describe() #Command for table showing  mean, standard deviation, min, max, and 25/50/75% percentiles for each feature and the target variables.
```

# Question 2.

```python
corr_mat = whitedf.corr() # Command for correlation matrix of the entire dataframe.
corr_mat
```

```python
sns.heatmap(corr_mat, annot=True, cmap='seismic', fmt='.2f', linewidths=0)
plt.title("Heatmap of the correlation matrix of entire white wine DataFrame")
plt.show();
```

```python
qual_corr_mat = whitedf.corr()["quality"] # Commands to create correlation nmatrix of target (quality) with the all feature variables.
qual_corr_mat
```

```python
target_corr = qual_corr_mat.reset_index()    # Barplot showing correlation of target (quality) with feature variables.
target_corr.columns = ['White wine quality', 'Correlation with Target']

sns.barplot(x='White wine quality', y='Correlation with Target', data= target_corr)
plt.xticks(rotation=65)
plt.title('Barplot showing correlation of each feature with target(quality)')
plt.show()
```

# Question 3.

```python
sns.histplot(data = whitedf,x='quality') #Commands creating histogram of target(quality)
plt.xlabel('Wine Quality')
plt.title("Histogram depicting White wine quality");
```

# Question 4

```python
colors = sns.color_palette("husl", len(whitedf.columns)) #this variable is then used to ensure each plot gets different color
for i, j in enumerate(whitedf.columns):                     #Loop used to create individual violin plots of all features
    sns.violinplot(x=whitedf[j], color=colors[i], inner= 'quart')
    plt.title(f'Violin plot of {j}', fontsize= 18)
    plt.xlabel(f'{j}')
    plt.show();
```

# Question 5.

```python
model = LinearRegression()    #Linear regression model training

model.fit(whitedf[['alcohol']],whitedf['quality'])

print(f'Intercept (b0) = {model.intercept_}')
print(f'Coefficient (b1) = {model.coef_[0]}')
```

```python
fitted = model.predict(whitedf[['alcohol']])      #model prediction

RMSE = np.sqrt(mean_squared_error(whitedf['quality'],fitted))
print(f'RMSE = {RMSE}')

r2 = r2_score(whitedf['quality'],fitted)
print(f'R2 = {r2}')
```

```python
residuals = whitedf['quality'] - fitted

# Create a Q-Q plot of the residuals for above trained model
stats.probplot(residuals, dist="norm", plot= plt)
plt.title("Q-Q plot of residuals")
plt.show()
```

```python
sns.relplot(x= 'alcohol' , y='quality', data= whitedf, color='darkred', label='Data Points')

# Plot regression line with a different color
plt.plot(whitedf['alcohol'], fitted , color='darkblue', linewidth= 3.50, label='Regression Line')
plt.xlabel('Alcohol')
plt.ylabel('Quality')                                                   #Commands creating scatterplot of obtained regression model and all data points of target and feature
plt.title('Alcohol vs. Quality with Linear Regression Model')
plt.legend()
plt.show()
```

```python
a = whitedf[['volatile acidity', 'alcohol', 'density']]
model = LinearRegression()
model.fit(a, whitedf['quality'])

m_fitted = model.predict(a)
RMSE = np.sqrt(mean_squared_error(whitedf['quality'], m_fitted)) #Commands training linear regression model with multiple variables
r2 = r2_score(whitedf['quality'], m_fitted)
print(f'Intercept (b0) = {model.intercept_}')
print(f'Coefficients (b1, b2, b3) = {model.coef_}')

print(f'RMSE: {RMSE}')
print(f'R2: {r2}')
```
